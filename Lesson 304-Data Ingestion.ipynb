**Using SparkSQL methods**

Using read API's

```python
%%sql
SELECT * FROM csv.`Files/Landing/`
```

Using temporary views

```python
%%sql
CREATE OR REPLACE TEMPORARY VIEW  TVW_Sales
AS 
SELECT _c0 As Date,_c1 As Country,_c2 As Units,_c3 As Revenue FROM csv.`Files/Landing/` WHERE _c1 IS NOT NULL and _c1 <>'Country'
```

```python
%%sql
select * from TVW_Sales
```

```python
%%sql
CREATE OR REPLACE TABLE Sales
USING DELTA
AS SELECT * FROM TVW_Sales
```

```python
%%sql
SELECT * FROM Sales
```

Alternative SparSQL option

```python
%%sql
INSERT INTO Sales
SELECT * FROM TVW_Sales
```

**Using PySpark methods**

```python
df=spark.read.format('csv').load('Files/Landing/')
display(df)
```

```python
from pyspark.sql.functions import col
df=df.where("_c1 IS NOT NULL and _c1 <>'Country'").select(col('_c0').alias('Date'),col('_c1').alias('Country'),col('_c2').alias('Units'),col('_c3').alias('Revenue'))
display(df)
```

```python
df.write.mode('overwrite').saveAsTable('Sales')
```

```python
%%sql
select count(*) from Sales
```

```python
df.write.mode('append').saveAsTable('Sales')
```

Alternative approach

```python
df.write.mode('append').save('Tables/Sales')
```
